```python
class TiendaOnline:
    def __init__(self):    # Inicializa los atributos principales de la tienda
        self.inventario = []  # Lista de productos dispo
        self.clientes = {}  # Registro de clientes
        self.ventas_totales = 0.0  # Ventas totales acumuladas
        self.carrito_de_compras = {}  # Carrito de compras del cliente

    def agregar_producto(self, nombre, precio, cantidad):
        #Agrega un producto al inventario o actualiza la cantidad si ya existe.
        
        for producto in self.inventario: # Itera sobre el inventario para buscar el producto
            if producto["nombre"].capitalize() == nombre.capitalize():
                 # Si el producto ya existe, actualizamos su cantidad
                producto["cantidad"] += cantidad
                print(f"Producto '{nombre.capitalize()}' actualizado. Nueva cantidad: {producto['cantidad']}")
                return # Salimos del método porque el producto ya fue actualizado
        # Si el producto no existe, lo añadimos como nuevo al inventario
        self.inventario.append({"nombre": nombre.capitalize(), "precio": precio, "cantidad": cantidad})
        print(f"Producto '{nombre.capitalize()}' añadido al inventario.")

    def ver_inventario(self):
        #Muestra el inventario de productos.
        
        if not self.inventario: # Si el inventario está vacío
            print("El inventario está vacío.")
        for producto in self.inventario: # Iteramos sobre cada producto en el inventario
            # Mostramos los detalles de cada producto
            print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")

    def buscar_producto(self, nombre):
        #Busca un producto en el inventario por nombre.
        
        for producto in self.inventario: # Iteramos sobre los productos del inventario
            if producto["nombre"].capitalize() == nombre.capitalize():
                # Si el producto existe, mostramos su información
                print(f"Producto encontrado: Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
                return # Salimos del método porque encontramos el producto
        # Si no se encuentra, mostramos un mensaje
        print(f"El producto '{nombre.capitalize()}' no se encuentra en el inventario.")

    def actualizar_stock(self, nombre, cantidad):
        #Actualiza el stock de un producto en el inventario.
        
        for producto in self.inventario: # Iteramos sobre los productos del inventario
            if producto["nombre"].capitalize() == nombre.capitalize():
                # Si el producto existe, actualizamos su cantidad
                producto["cantidad"] += cantidad
                print(f"El stock del producto '{nombre.capitalize()}' ha sido actualizado. Nueva cantidad: {producto['cantidad']}")
                return
            # Si el producto no existe, mostramos un mensaje
        print(f"El producto '{nombre.capitalize()}' no se encuentra en el inventario.")

    def eliminar_producto(self, nombre):
        #Elimina un producto del inventario.
        for producto in self.inventario: # Iteramos sobre los productos del inventario
            if producto["nombre"].capitalize() == nombre.capitalize():
                # Si el producto existe, lo eliminamos 
                self.inventario.remove(producto)
                print(f"El producto '{nombre.capitalize()}' ha sido eliminado del inventario.")
                return
            # Si el producto no existe, mostramos un mensaje
        print(f"El producto '{nombre.capitalize()}' no se encuentra en el inventario.")

    def calcular_valor_inventario(self):
        #Calcula y muestra el valor total del inventario.
        valor_total = 0 # Variable para acumular el valor total
        for producto in self.inventario: # Iteramos sobre los productos del inventario
            # Calculamos el valor total sumando el precio por la cantidad de cada producto
            valor_total += producto["precio"] * producto["cantidad"]
             # Mostramos el valor total
        print(f"El valor total del inventario es: ${valor_total}")
        return valor_total    # Retornamos el valor total
    

    def realizar_compra(self):  # Permite al cliente realizar una compra seleccionando productos del inventario
        
        carrito_de_compras = {} # Inicializamos un carrito vacío
        total = 0  # Inicializamos el total de la compra en 0
        
        while True: # Iniciamos un bucle para permitir múltiples selecciones
            self.ver_inventario()   # Mostramos el inventario disponible
            comprar_producto = input("Dime el nombre del producto (o 'Salir' para finalizar): ").capitalize()
            if comprar_producto == "Salir":  # Si el cliente quiere salir del bucle
                break  # Salimos del bucle

            cantidad_producto = int(input(f"Dime cuántas unidades de '{comprar_producto}' quieres comprar: "))

            for producto in self.inventario: # Buscamos el producto en el inventario
                if producto["nombre"].capitalize() == comprar_producto:
                    if cantidad_producto > producto["cantidad"]: # Si no hay suficiente stock
                        print(f"No tenemos suficiente stock. Solo hay {producto['cantidad']} disponible.")
                    else:  
                         # Reducimos la cantidad en el inventario
                        producto["cantidad"] -= cantidad_producto
                        print(f'El stock del {producto["nombre"]} ha sido actualizado a {producto["cantidad"]}')

                        # Usamos .update() para actualizar el carrito
                        if comprar_producto in carrito_de_compras:
                            # Si el producto ya está en el carrito, sumamos la cantidad
                            carrito_de_compras.update({comprar_producto: carrito_de_compras[comprar_producto] + cantidad_producto})
                            total += producto ['precio'] * cantidad_producto
                        else:
                            # Si no está en el carrito, lo añadimos
                            carrito_de_compras.update({comprar_producto: cantidad_producto})
                            print(carrito_de_compras)

                             # Actualizamos el total de la compra
                            total += producto ['precio'] * cantidad_producto
                            print(f"{cantidad_producto} unidad(es) de '{comprar_producto}' añadida(s) al carrito.")
                 
            else:
                # Si el producto no está en el inventario
                print(f"El producto '{comprar_producto}' no está en el inventario.")

             # Mostramos el total acumulado del carrito
            print(f'El valor del carrito es {total}€.')
               
```

```python
# la tienda
Grafos = TiendaOnline()

# Agregar productos al inventario
Grafos.agregar_producto("Camisa", 25.00, 10)
Grafos.agregar_producto("Pantalón", 40.00, 5)
Grafos.agregar_producto("Calcetines", 5.00, 20)
Grafos.agregar_producto("Camiseta", 15.00, 12)

# Verificar 
Grafos.ver_inventario()
```
